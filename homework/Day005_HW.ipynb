```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r.text)
print('ä¸€æ¬¡æœƒå›å‚³'+str(len(data))+'ç­†è³‡æ–™')
print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«çš„æ¬„ä½æœ‰:')
for d in data[0]:
    print(d)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
    print('')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

r_pop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data_pop = json.loads(r_pop.text)

r_notPop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data_notPop = json.loads(r_notPop.text)

pop_avg_commentCount = 0
pop_avg_likeCount = 0
notPop_avg_commentCount = 0
notPop_avg_likeCount = 0

#ç†±é–€æ–‡ç« 
for d in data_pop:
    pop_avg_commentCount += d['commentCount']
    pop_avg_likeCount += d['likeCount']
pop_avg_commentCount /=30
pop_avg_likeCount /=30
print('ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸:'+str(round(pop_avg_commentCount)))
print('ç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸:'+str(round(pop_avg_likeCount)))

#éç†±é–€æ–‡ç« 
for d in data_notPop:
    notPop_avg_commentCount += d['commentCount']
    notPop_avg_likeCount += d['likeCount']
notPop_avg_commentCount /=30
notPop_avg_likeCount /=30
print('éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸:'+str(round(notPop_avg_commentCount)))
print('éç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸:'+str(round(notPop_avg_likeCount)))
```
