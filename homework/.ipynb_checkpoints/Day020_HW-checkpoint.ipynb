```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
# 將資料存進一個陣列
temp = []
for tr in table.find_all('tr'):
    for td in tr.find_all('td'):
        temp.append(td.getText())

# 將陣列中值為'\xa0'(不間斷空白符)刪除
for i in temp:
    if (i=='\xa0'):
        temp.remove(i)

# 轉為二維陣列
import numpy as np 
data = np.array(temp).reshape(10,26)

# 設定成字典型態
res = {}
for i in range(0,10):
    for count,j in enumerate(data[i]):
        if(count==0):
            title = j
            res.setdefault(title,{})
            continue
        if(count==1):
            continue
        if(count%2==0):
            date = j
        if(count%2!=0):
            num = j
            res[title][date]=num
            
print(res)
```
