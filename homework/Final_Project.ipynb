# 基本目標

```python
import requests
import json
import pandas as pd
```

```python
# 透過API爬取熱門新聞內容，並加資料存為字典型態
news = []
for n in range(30, 540, 30): 
    url = 'https://www.cupoy.com/MixNewsMongoAction.do?op=getTopMixNews&nationid=TW&startNewsID={}&len=30'.format(n)
    response = requests.get(url)
    data = json.loads(response.text)
    for d in data['items']:
        if 'bucketgrpids' in d:
            news.append({
                'title': d['title'],
                'bucketgrpids':d['bucketgrpids'][0]
            })
```

```python
# 表格呈現
data=pd.DataFrame(news)
data
```

```python
#根據類別的出現頻率做一個升降排序
data_group=pd.DataFrame(data.groupby('bucketgrpids').size(),columns=['amounts'])
data_group=data_group.sort_values(by='amounts',ascending=False)
data_amount=data_group.reset_index()
data_amount
```

```python
#讓plt可以呈現中文
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(18,20)) #圖形大小

labels = data_amount['bucketgrpids'][:10] #種類名稱
size = data_amount['amounts'][:10] #數據
explode = [0.15,0.1,0.05,0,0,0,0,0,0,0] #選擇哪些要突出
plt.pie(size,explode= explode, labels = labels, autopct= "%2.1f%%",textprops = {"fontsize":30},shadow=True,startangle=360)
#autopct= "%0.1f%%"小數點前=前面顯示幾格，小數點後顯示幾個數字
plt.title("Cupoy 新聞前500篇的熱門文章種類", {"fontsize" : 40})  # 設定標題及其文字大小
plt.legend(loc = 1,fontsize = 'large') #示意圖的位置，字體大小
plt.show()
```

## 斷詞+過濾stop words+計算同樣文字出現的頻率且排名

```python
import jieba
import re
import nltk
jieba.set_dictionary("./dict.txt.big.txt")
```

```python
# 自己設定了幾個，因為他把他們分開看XD
jieba.suggest_freq(('風傳媒'),True)
jieba.suggest_freq(('商周財富網'),True)
jieba.suggest_freq(('ETtoday旅遊雲'),True)
jieba.suggest_freq(('關鍵評論網'),True)
jieba.suggest_freq(('經理人'),True)
jieba.suggest_freq(('姊妹淘'),True)
jieba.suggest_freq(('ETtoday旅遊新聞旅遊'),True)
```

```python
# 排除標點符號
title = [re.sub("\W+","",n['title']).replace("_","") for n in news]
title[:100]
```

```python
# 用來判斷是否為數字
def is_number(num):
    pattern = re.compile(r'\d+')
    result = pattern.match(num)
    if result:
        return True
    else:
        return False
```

```python
# 計算tokenization後，每個字的數量，排序後印出(print)前100最大數量的詞，然後"肉眼看"那些詞應該被去除，把那些該去除的詞放入stop_words
tok =[word for sen in title for word in jieba.cut(sen,HMM=False) if is_number(word) == False]
fdlist = nltk.FreqDist(tok).most_common()
print (fdlist[:50])
```

```python
# 把那些該去除的詞放入stop_words
stop_words = ["的","你","超","為","大","車","是","有","小","什麼","不","訊","了","在","都","與","上","好","我","就","中","能","將","之","人","心","讓","看","網","個","到","新"]
```

```python
# 利用stop_words清理資料，回傳清理過後的資料
chinese_stopped = [word for sen in title for word in jieba.cut(sen,HMM=False) if is_number(word) == False and word not in stop_words] 
fdlist = nltk.FreqDist(chinese_stopped).most_common()
print (fdlist[:50])
```

# 製作文字雲

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 將關鍵字放入List
data = [f[0] for f in fdlist]
wl_space_split = " ".join(data)
```

```python
# 設定文字雲格式
wc = WordCloud(background_color='white',width=1000, height=600, margin=2,font_path= r'msjh.ttc',random_state=None)
wc.generate(wl_space_split)
plt.figure(figsize=(20,20))
plt.imshow(wc,interpolation='bilinear')
plt.axis("off")
plt.show()
```
