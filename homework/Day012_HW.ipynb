## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 取出今天所有的發文

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
data = []

# 取得所有分類標籤
new_tag = []
tag = ''
for d in soup.find(class_='part_list_2').find_all('h3'):
    for n in new_tag:
        if(n == d.find('em').text):
            tag =d.find('em').text
            break
    if(tag != d.find('em').text):
        tag =d.find('em').text
        new_tag.append(tag)
        
# 建立標籤字典
for n in new_tag:
    news = {}
    news['tag'] = n
    news['content'] = []
    data.append(news)
    
# 存內容
content = []
for d in soup.find(class_='part_list_2').find_all('h3'):
    news = {}
    news['data'] = d.find(class_='date').text
    news['title'] = d.find('a').text
    for t in data:
        if(d.find('em').text==t['tag']):
            t['content'].append(news)
print(data)
```

### ③ 哪一個類別的文章最多

```python
max = ''
num = 0
for d in data:
    if(len(d['content'])>num):
        num = len(d['content'])
        max = d['tag']             
print(max)
```
